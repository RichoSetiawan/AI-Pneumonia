```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from tensorflow import keras
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications import resnet50, densenet, inception_v3
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries

# ============================================================================
# BAGIAN 1: FUNGSI UTILITAS
# ============================================================================

def load_trained_models(model_paths_dict):
    """Memuat model-model Keras dari path yang diberikan."""
    loaded_models = {}
    for name, path in model_paths_dict.items():
        if not os.path.exists(path):
            print(f" Peringatan: File model '{name}' tidak ditemukan di: {path}")
            continue
        try:
            print(f"  - Memuat model '{name}'...")
            model = keras.models.load_model(path)
            loaded_models[name] = model
            print(f"   Model '{name}' berhasil dimuat.")
        except Exception as e:
            print(f"   Gagal memuat model '{name}': {e}")
    return loaded_models

def get_preprocessor(model_name):
    """Mengembalikan fungsi preprocessing berdasarkan nama model."""
    preprocessors = {
        'ResNet50': resnet50.preprocess_input,
        'DenseNet121': densenet.preprocess_input,
        'InceptionV3': inception_v3.preprocess_input
    }
    if model_name not in preprocessors:
        raise ValueError(f"Preprocessor tidak dikenal untuk model: {model_name}")
    return preprocessors[model_name]

def get_image_size(model_name):
    """Mengembalikan ukuran gambar yang sesuai untuk model."""
    sizes = {
        'ResNet50': (224, 224),
        'DenseNet121': (224, 224),
        'InceptionV3': (299, 299)
    }
    return sizes.get(model_name, (224, 224))

# ============================================================================
# BAGIAN 2: PREDIKSI MODEL INDIVIDU
# ============================================================================

def predict_single_model(image_path, model_name, model):
    """
    Melakukan prediksi dengan satu model.
    
    Returns:
        dict: {'model': nama_model, 'prediction': kelas, 'probability': probabilitas}
    """
    print(f"\n--- Prediksi Model {model_name} ---")
    
    try:
        img_size = get_image_size(model_name)
        preprocess_func = get_preprocessor(model_name)
        
        # Muat dan proses gambar
        img = load_img(image_path, target_size=img_size)
        img_array = img_to_array(img)
        processed_array = preprocess_func(np.expand_dims(img_array, axis=0))
        
        # Prediksi
        pred_proba = model.predict(processed_array, verbose=0)[0, 0]
        pred_class = "Pneumonia" if pred_proba > 0.5 else "Normal"
        
        print(f"  Probabilitas Pneumonia: {pred_proba:.4f}")
        print(f"  Prediksi: {pred_class}")
        
        return {
            'model': model_name,
            'prediction': pred_class,
            'probability': float(pred_proba)
        }
    except Exception as e:
        print(f"   Error: {e}")
        return None

def predict_all_models(image_path, models_dict):
    """Melakukan prediksi dengan semua model individu."""
    print(f"\n{'='*70}")
    print(f" PREDIKSI MODEL INDIVIDU")
    print(f"{'='*70}")
    
    results = {}
    for model_name, model in models_dict.items():
        result = predict_single_model(image_path, model_name, model)
        if result:
            results[model_name] = result
    
    return results

# ============================================================================
# BAGIAN 3: PREDIKSI ENSEMBLE
# ============================================================================

def predict_ensemble(image_path, models_dict, ensemble_weights=None):
    """
    Melakukan prediksi ensemble dari semua model.
    
    Returns:
        dict: {'prediction': kelas, 'probability': probabilitas, 'individual_predictions': dict}
    """
    print(f"\n{'='*70}")
    print(f" PREDIKSI ENSEMBLE")
    print(f"{'='*70}")
    
    if ensemble_weights is None:
        ensemble_weights = [1.0 / len(models_dict)] * len(models_dict)
    
    model_predictions = []
    individual_results = {}
    
    for model_name, model in models_dict.items():
        try:
            img_size = get_image_size(model_name)
            preprocess_func = get_preprocessor(model_name)
            
            img = load_img(image_path, target_size=img_size)
            img_array = img_to_array(img)
            processed_array = preprocess_func(np.expand_dims(img_array, axis=0))
            
            pred_proba = model.predict(processed_array, verbose=0)[0, 0]
            model_predictions.append(pred_proba)
            individual_results[model_name] = float(pred_proba)
            
            print(f"  {model_name}: {pred_proba:.4f}")
            
        except Exception as e:
            print(f"   Error pada {model_name}: {e}")
    
    if not model_predictions:
        print(" Tidak ada prediksi yang berhasil")
        return None
    
    # Hitung ensemble
    ensemble_proba = np.average(model_predictions, weights=ensemble_weights[:len(model_predictions)])
    ensemble_class = "Pneumonia" if ensemble_proba > 0.5 else "Normal"
    
    print(f"\n  Probabilitas Ensemble: {ensemble_proba:.4f}")
    print(f"  Prediksi Ensemble: {ensemble_class}")
    
    return {
        'prediction': ensemble_class,
        'probability': float(ensemble_proba),
        'individual_predictions': individual_results
    }

# ============================================================================
# BAGIAN 4: LIME UNTUK MODEL INDIVIDU
# ============================================================================

def lime_single_model(image_path, model_name, model, class_names=['Normal', 'Pneumonia']):
    """
    Menghasilkan analisis LIME untuk satu model.
    """
    print(f"\n{'='*70}")
    print(f" LIME ANALISIS - {model_name}")
    print(f"{'='*70}")
    
    try:
        img_size = get_image_size(model_name)
        preprocess_func = get_preprocessor(model_name)
        
        # Muat gambar
        img = load_img(image_path, target_size=img_size)
        img_array = img_to_array(img)
        img_array_uint8 = np.uint8(img_array.copy())
        
        # Fungsi prediksi untuk LIME
        def predict_fn(images):
            processed_batch = []
            for img in images:
                img_resized = cv2.resize(img.copy(), img_size)
                img_expanded = np.expand_dims(img_resized, axis=0)
                img_processed = preprocess_func(img_expanded)
                processed_batch.append(img_processed[0])
            
            batch_array = np.array(processed_batch)
            preds = model.predict(batch_array, verbose=0)
            probs_pneumonia = preds[:, 0]
            return np.column_stack((1 - probs_pneumonia, probs_pneumonia))
        
        # Inisialisasi LIME
        explainer = lime_image.LimeImageExplainer(random_state=42)
        
        # Generate explanation
        print(f"  Menghasilkan penjelasan LIME...")
        explanation = explainer.explain_instance(
            img_array_uint8,
            predict_fn,
            top_labels=2,
            hide_color=0,
            num_samples=1000,
            batch_size=128,
            random_seed=42
        )
        
        # Dapatkan prediksi
        pred_proba = predict_fn(np.expand_dims(img_array_uint8, axis=0))[0]
        pred_class_idx = 1 if pred_proba[1] > 0.5 else 0
        pred_class = class_names[pred_class_idx]
        
        # Ambil mask untuk kelas yang diprediksi
        try:
            temp, mask = explanation.get_image_and_mask(
                pred_class_idx,
                positive_only=False,
                num_features=10,
                hide_rest=False,
                min_weight=0.01
            )
        except:
            # Fallback jika kelas tidak tersedia
            available_labels = explanation.top_labels
            pred_class_idx = available_labels[0] if available_labels else 1
            temp, mask = explanation.get_image_and_mask(
                pred_class_idx,
                positive_only=False,
                num_features=10,
                hide_rest=False
            )
        
        # Visualisasi
        fig, axes = plt.subplots(1, 4, figsize=(24, 6))
        fig.suptitle(
            f'LIME Analysis - {model_name}\n'
            f'Prediksi: {pred_class} (Prob: {pred_proba[1]:.4f})',
            fontsize=14, fontweight='bold'
        )
        
        # Gambar asli
        axes[0].imshow(img_array_uint8.astype('uint8'))
        axes[0].set_title('Gambar Asli', fontsize=12, fontweight='bold')
        axes[0].axis('off')
        
        # LIME boundaries
        img_bound = mark_boundaries(temp / 255.0, mask)
        axes[1].imshow(img_bound)
        axes[1].set_title('LIME Boundaries', fontsize=12, fontweight='bold')
        axes[1].axis('off')
        
        # LIME mask
        im = axes[2].imshow(mask, cmap='RdBu_r', vmin=-1, vmax=1)
        axes[2].set_title(f'LIME Mask\n(Merah = Mendukung {class_names[pred_class_idx]})', 
                         fontsize=12, fontweight='bold')
        axes[2].axis('off')
        plt.colorbar(im, ax=axes[2], fraction=0.046, pad=0.04)
        
        # Heatmap kontribusi
        heatmap = np.zeros_like(mask, dtype=float)
        for seg_id in np.unique(mask):
            if seg_id == -1:
                continue
            seg_mask = (mask == seg_id)
            img_with_seg = img_array_uint8.copy()
            img_with_seg[~seg_mask] = 0
            
            prob_with_seg = predict_fn(np.expand_dims(img_with_seg, axis=0))[0]
            contribution = prob_with_seg[pred_class_idx] - pred_proba[pred_class_idx]
            heatmap[seg_mask] = contribution
        
        im2 = axes[3].imshow(heatmap, cmap='seismic', vmin=-0.3, vmax=0.3)
        axes[3].set_title('Heatmap Kontribusi', fontsize=12, fontweight='bold')
        axes[3].axis('off')
        plt.colorbar(im2, ax=axes[3], fraction=0.046, pad=0.04)
        
        plt.tight_layout(rect=[0, 0, 1, 0.93])
        
        output_filename = f'lime_{model_name}_{os.path.basename(image_path)}.png'
        plt.savefig(output_filename, dpi=150, bbox_inches='tight')
        print(f"   Visualisasi disimpan: {output_filename}")
        plt.show()
        
        return {
            'model': model_name,
            'prediction': pred_class,
            'probability': float(pred_proba[1]),
            'success': True
        }
        
    except Exception as e:
        print(f"   Error: {e}")
        import traceback
        traceback.print_exc()
        return None

def lime_all_models(image_path, models_dict):
    """Menghasilkan analisis LIME untuk semua model individu."""
    print(f"\n{'='*70}")
    print(f" LIME ANALISIS - SEMUA MODEL INDIVIDU")
    print(f"{'='*70}")
    
    results = {}
    for model_name, model in models_dict.items():
        result = lime_single_model(image_path, model_name, model)
        if result:
            results[model_name] = result
    
    return results

# ============================================================================
# BAGIAN 5: LIME ENSEMBLE
# ============================================================================

def lime_ensemble(image_path, models_dict, ensemble_weights=None, class_names=['Normal', 'Pneumonia']):
    """
    Menghasilkan analisis LIME untuk ensemble model.
    """
    print(f"\n{'='*70}")
    print(f" LIME ANALISIS - ENSEMBLE")
    print(f"{'='*70}")
    
    if ensemble_weights is None:
        ensemble_weights = [1.0 / len(models_dict)] * len(models_dict)
    
    try:
        # Muat gambar dengan ukuran default
        img_size = (224, 224)
        img = load_img(image_path, target_size=img_size)
        img_array = img_to_array(img)
        img_array_uint8 = np.uint8(img_array.copy())
        
        # Resize jika terlalu besar
        MAX_SIZE = 512
        if img_array_uint8.shape[0] > MAX_SIZE or img_array_uint8.shape[1] > MAX_SIZE:
            scale = MAX_SIZE / max(img_array_uint8.shape[0], img_array_uint8.shape[1])
            new_size = (int(img_array_uint8.shape[1] * scale), int(img_array_uint8.shape[0] * scale))
            img_array_uint8 = cv2.resize(img_array_uint8, new_size)
        
        # Fungsi prediksi ensemble
        def ensemble_predict_fn(images):
            ensemble_predictions = []
            
            for model_name, model in models_dict.items():
                img_size_model = get_image_size(model_name)
                preprocess_func = get_preprocessor(model_name)
                
                processed_batch = []
                for img in images:
                    img_resized = cv2.resize(img.copy(), img_size_model)
                    img_expanded = np.expand_dims(img_resized, axis=0)
                    img_processed = preprocess_func(img_expanded)
                    processed_batch.append(img_processed[0])
                
                batch_array = np.array(processed_batch)
                model_preds = model.predict(batch_array, verbose=0)
                probs_pneumonia = model_preds[:, 0]
                ensemble_predictions.append(probs_pneumonia)
            
            ensemble_probas = np.average(ensemble_predictions, axis=0, weights=ensemble_weights)
            return np.column_stack((1 - ensemble_probas, ensemble_probas))
        
        # Inisialisasi LIME
        explainer = lime_image.LimeImageExplainer(random_state=42)
        
        # Generate explanation
        print(f"  Menghasilkan penjelasan LIME ensemble...")
        explanation = explainer.explain_instance(
            img_array_uint8,
            ensemble_predict_fn,
            top_labels=2,
            hide_color=0,
            num_samples=1000,
            batch_size=128,
            random_seed=42
        )
        
        # Dapatkan prediksi
        pred_proba = ensemble_predict_fn(np.expand_dims(img_array_uint8, axis=0))[0]
        pred_class_idx = 1 if pred_proba[1] > 0.5 else 0
        pred_class = class_names[pred_class_idx]
        
        # Ambil mask
        try:
            temp, mask = explanation.get_image_and_mask(
                pred_class_idx,
                positive_only=False,
                num_features=10,
                hide_rest=False,
                min_weight=0.01
            )
        except:
            available_labels = explanation.top_labels
            pred_class_idx = available_labels[0] if available_labels else 1
            temp, mask = explanation.get_image_and_mask(
                pred_class_idx,
                positive_only=False,
                num_features=10,
                hide_rest=False
            )
        
        # Visualisasi
        fig, axes = plt.subplots(1, 4, figsize=(24, 6))
        fig.suptitle(
            f'LIME Analysis - ENSEMBLE\n'
            f'Prediksi: {pred_class} (Prob: {pred_proba[1]:.4f})',
            fontsize=14, fontweight='bold'
        )
        
        # Gambar asli
        axes[0].imshow(img_array_uint8.astype('uint8'))
        axes[0].set_title('Gambar Asli', fontsize=12, fontweight='bold')
        axes[0].axis('off')
        
        # LIME boundaries
        img_bound = mark_boundaries(temp / 255.0, mask)
        axes[1].imshow(img_bound)
        axes[1].set_title('LIME Boundaries', fontsize=12, fontweight='bold')
        axes[1].axis('off')
        
        # LIME mask
        im = axes[2].imshow(mask, cmap='RdBu_r', vmin=-1, vmax=1)
        axes[2].set_title(f'LIME Mask\n(Merah = Mendukung {class_names[pred_class_idx]})', 
                         fontsize=12, fontweight='bold')
        axes[2].axis('off')
        plt.colorbar(im, ax=axes[2], fraction=0.046, pad=0.04)
        
        # Heatmap kontribusi
        heatmap = np.zeros_like(mask, dtype=float)
        for seg_id in np.unique(mask):
            if seg_id == -1:
                continue
            seg_mask = (mask == seg_id)
            img_with_seg = img_array_uint8.copy()
            img_with_seg[~seg_mask] = 0
            
            prob_with_seg = ensemble_predict_fn(np.expand_dims(img_with_seg, axis=0))[0]
            contribution = prob_with_seg[pred_class_idx] - pred_proba[pred_class_idx]
            heatmap[seg_mask] = contribution
        
        im2 = axes[3].imshow(heatmap, cmap='seismic', vmin=-0.3, vmax=0.3)
        axes[3].set_title('Heatmap Kontribusi', fontsize=12, fontweight='bold')
        axes[3].axis('off')
        plt.colorbar(im2, ax=axes[3], fraction=0.046, pad=0.04)
        
        plt.tight_layout(rect=[0, 0, 1, 0.93])
        
        output_filename = f'lime_ensemble_{os.path.basename(image_path)}.png'
        plt.savefig(output_filename, dpi=150, bbox_inches='tight')
        print(f"   Visualisasi disimpan: {output_filename}")
        plt.show()
        
        return {
            'model': 'Ensemble',
            'prediction': pred_class,
            'probability': float(pred_proba[1]),
            'success': True
        }
        
    except Exception as e:
        print(f"   Error: {e}")
        import traceback
        traceback.print_exc()
        return None

# ============================================================================
# BAGIAN 6: FUNGSI UTAMA
# ============================================================================

def analyze_image(image_path, models_dict):
    """
    Fungsi utama untuk menganalisis satu gambar dengan lengkap:
    1. Prediksi model individu
    2. Prediksi ensemble
    3. LIME model individu
    4. LIME ensemble
    """
    print(f"\n{'='*80}")
    print(f" ANALISIS LENGKAP GAMBAR")
    print(f"{'='*80}")
    print(f"Gambar: {os.path.basename(image_path)}")
    print(f"Path: {image_path}")
    
    if not os.path.exists(image_path):
        print(f" ERROR: Gambar tidak ditemukan!")
        return None
    
    results = {
        'image_path': image_path,
        'filename': os.path.basename(image_path)
    }
    
    # 1. Prediksi model individu
    individual_preds = predict_all_models(image_path, models_dict)
    results['individual_predictions'] = individual_preds
    
    # 2. Prediksi ensemble
    ensemble_pred = predict_ensemble(image_path, models_dict)
    results['ensemble_prediction'] = ensemble_pred
    
    # 3. LIME model individu
    individual_lime = lime_all_models(image_path, models_dict)
    results['individual_lime'] = individual_lime
    
    # 4. LIME ensemble
    ensemble_lime = lime_ensemble(image_path, models_dict)
    results['ensemble_lime'] = ensemble_lime
    
    # Ringkasan
    print(f"\n{'='*80}")
    print(f" RINGKASAN HASIL")
    print(f"{'='*80}")
    print(f"Gambar: {os.path.basename(image_path)}\n")
    
    print("PREDIKSI MODEL INDIVIDU:")
    for model_name, pred in individual_preds.items():
        print(f"  {model_name:15s}: {pred['prediction']:10s} (Prob: {pred['probability']:.4f})")
    
    if ensemble_pred:
        print(f"\nPREDIKSI ENSEMBLE:")
        print(f"  {'Ensemble':15s}: {ensemble_pred['prediction']:10s} (Prob: {ensemble_pred['probability']:.4f})")
    
    print(f"\n{'='*80}")
    
    return results

# ============================================================================
# BAGIAN 7: EKSEKUSI UTAMA
# ============================================================================

if __name__ == "__main__":
    # Definisikan path model
    model_paths = {
        'ResNet50': '/kaggle/input/pneumonia-84/keras/default/1/ResNet50.keras',
        'DenseNet121': '/kaggle/input/pneumonia-92/keras/default/1/densenet121_pneumonia.keras',
        'InceptionV3': '/kaggle/input/pneumonia-82/keras/default/1/inceptionv3_full_model.keras',
    }
    
    # Muat model
    print("="*80)
    print("MEMUAT MODEL")
    print("="*80)
    trained_models = load_trained_models(model_paths)
    
    if not trained_models:
        print(" Tidak ada model yang berhasil dimuat!")
    else:
        print(f"\n {len(trained_models)} model berhasil dimuat: {list(trained_models.keys())}")
        
        # PATH GAMBAR YANG INGIN DIANALISIS
        # Ganti path ini sesuai kebutuhan Anda
        image_path = '/kaggle/input/pneumonia/Pneumonia/test/NORMAL/NORMAL2-IM-0302-0001.jpeg'
        
        # Jalankan analisis lengkap
        results = analyze_image(image_path, trained_models)
        
        print(f"\n{'='*80}")
        print(" ANALISIS SELESAI!")
        print(f"{'='*80}")
        print("File visualisasi telah disimpan:")
        print("  - lime_ResNet50_*.png")
        print("  - lime_DenseNet121_*.png")
        print("  - lime_InceptionV3_*.png")
        print("  - lime_ensemble_*.png")
        print(f"{'='*80}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 # Diperlukan untuk resize dalam fungsi LIME
from tensorflow import keras
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications import resnet50, densenet, inception_v3
# Impor library XAI (LIME)
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries

def load_trained_models(model_paths_dict):
    """
    Memuat model-model Keras dari path-file yang diberikan.

    Args:
        model_paths_dict (dict): Kamus dengan format {'model_name': 'path/to/model.keras'}

    Returns:
        dict: Kamus berisi model-model yang telah dimuat {'model_name': model_object}
    """
    loaded_models = {}
    for name, path in model_paths_dict.items():
        if not os.path.exists(path):
            print(f" Peringatan: File model untuk '{name}' tidak ditemukan di: {path}")
            continue
        try:
            print(f"  - Memuat model '{name}' dari {path}...")
            model = keras.models.load_model(path)
            loaded_models[name] = model
            print(f"   Model '{name}' berhasil dimuat.")
        except Exception as e:
            print(f"   Gagal memuat model '{name}' dari {path}: {e}")
    return loaded_models

def predict_pneumonia_ensemble(image_path, models_dict, ensemble_weights=None):
    """
    Melakukan prediksi pneumonia/normal pada satu gambar menggunakan ensemble dari tiga model.

    Args:
        image_path (str): Path ke file gambar X-ray.
        models_dict (dict): Kamus berisi model-model yang telah dilatih {'model_name': model_object}.
        ensemble_weights (list, optional): Bobot untuk setiap model dalam ensemble. 
                                          Jika None, digunakan rata-rata sederhana [1/3, 1/3, 1/3].

    Returns:
        tuple: (predicted_class_label, ensemble_probability)
               predicted_class_label: 'Normal' atau 'Pneumonia'
               ensemble_probability: Probabilitas kelas 'Pneumonia' menurut ensemble (float antara 0 dan 1)
    """
    if ensemble_weights is None:
        ensemble_weights = [1.0 / len(models_dict)] * len(models_dict)

    if len(models_dict) != len(ensemble_weights):
        raise ValueError("Jumlah model harus sama dengan jumlah bobot ensemble.")

    # Siapkan fungsi preprocessing dan ukuran gambar untuk setiap arsitektur
    preprocess_funcs = {
        "ResNet50": resnet50.preprocess_input,
        "DenseNet121": densenet.preprocess_input,
        "InceptionV3": inception_v3.preprocess_input
    }
    img_sizes = {
        "ResNet50": (224, 224),
        "DenseNet121": (224, 224),
        "InceptionV3": (299, 299)
    }

    model_predictions = []
    model_names_used = []

    for model_name, model in models_dict.items():
        if model_name not in preprocess_funcs or model_name not in img_sizes:
            print(f" Peringatan: Konfigurasi untuk model '{model_name}' tidak ditemukan. Lewati.")
            continue

        try:
            print(f"  - Memproses gambar untuk model {model_name}...")
            # Muat dan proses gambar
            img = load_img(image_path, target_size=img_sizes[model_name])
            img_array = img_to_array(img)
            processed_array = preprocess_funcs[model_name](np.expand_dims(img_array, axis=0))

            # Prediksi
            pred_proba = model.predict(processed_array, verbose=0)[0, 0] # Ambil probabilitas pneumonia (kelas 1)
            model_predictions.append(pred_proba)
            model_names_used.append(model_name)
            print(f"    - Prediksi {model_name}: {pred_proba:.4f}")

        except Exception as e:
            print(f"   Gagal memproses gambar untuk model {model_name}: {e}")

    if not model_predictions:
        print(" Tidak ada prediksi yang berhasil diambil dari model-model yang dimuat.")
        return None, None

    # Pastikan bobot sesuai dengan model yang digunakan
    if len(model_predictions) != len(ensemble_weights):
         # Jika model gagal dimuat, sesuaikan bobot
         effective_weights = [w for i, w in enumerate(ensemble_weights) if list(models_dict.keys())[i] in model_names_used]
         if len(effective_weights) == 0:
             print(" Tidak ada bobot yang sesuai untuk model yang digunakan.")
             return None, None
         # Normalisasi ulang bobot jika jumlahnya berubah
         total_weight = sum(effective_weights)
         if total_weight > 0:
             effective_weights = [w / total_weight for w in effective_weights]
         else:
             effective_weights = [1.0 / len(model_predictions)] * len(model_predictions)
         ensemble_weights = effective_weights

    # Hitung rata-rata terbobot untuk prediksi ensemble
    ensemble_proba = np.average(model_predictions, weights=ensemble_weights)

    # Tentukan kelas akhir berdasarkan threshold 0.5
    final_prediction = "Pneumonia" if ensemble_proba > 0.5 else "Normal"

    print(f"\n--- Prediksi Ensemble ---")
    print(f"  - Gambar: {os.path.basename(image_path)}")
    print(f"  - Probabilitas Pneumonia (Ensemble): {ensemble_proba:.4f}")
    print(f"  - Prediksi Kelas: {final_prediction}")

    return final_prediction, ensemble_proba

def get_preprocessor(model_name):
    """Mengembalikan fungsi preprocessing yang benar berdasarkan nama model."""
    if model_name == 'ResNet50':
        return resnet50.preprocess_input
    elif model_name == 'DenseNet121':
        return densenet.preprocess_input
    elif model_name == 'InceptionV3':
        return inception_v3.preprocess_input
    else:
        raise ValueError(f"Preprocessor tidak dikenal untuk model: {model_name}")

def calculate_xai_lime_ensemble(models_dict, ensemble_weights, img_path, class_names=['Normal', 'Pneumonia']):
    """
    Fungsi inti yang menjalankan pipeline XAI (LIME) untuk satu gambar
    menggunakan prediksi ensemble aktif.
    """
    print(f"\n--- Menganalisis XAI (LIME) Ensemble untuk {os.path.basename(img_path)} ---")
    # --- 1. Persiapan Gambar & Model ---
    # Kita tetap perlu preprocessing untuk LIME, jadi kita pilih salah satu model untuk ukuran
    # Kita asumsikan ukuran gambar untuk ResNet50/DenseNet121 sebagai default jika bukan InceptionV3
    img_size = (299, 299) if any('InceptionV3' in name for name in models_dict.keys()) else (224, 224)
    # Kita pilih preprocessor dari model pertama sebagai default, atau tentukan secara spesifik jika diperlukan
    # Kita buat wrapper predict_fn untuk ensemble
    def ensemble_predict_fn(images):
        """
        Fungsi prediksi untuk LIME yang menerima batch gambar dan mengembalikan probabilitas ensemble.
        """
        ensemble_predictions = []
        for model_name, model in models_dict.items():
            img_size_model = (299, 299) if 'InceptionV3' in model_name else (224, 224)
            preprocess_func = get_preprocessor(model_name)
            # Resize gambar batch ke ukuran model yang sesuai sebelum preprocessing
            processed_images = np.array([cv2.resize(img, img_size_model) for img in images])
            processed_images = preprocess_func(processed_images)
            model_pred = model.predict(processed_images, verbose=0)
            ensemble_predictions.append(model_pred.flatten()) # Ambil probabilitas pneumonia

        # Gabungkan prediksi model dengan bobot ensemble
        ensemble_probas = np.average(ensemble_predictions, axis=0, weights=ensemble_weights)
        # LIME mengharapkan (num_samples, num_classes), jadi kita reshape
        prob_pos = ensemble_probas # Probabilitas kelas positif (Pneumonia)
        prob_neg = 1 - prob_pos
        return np.column_stack((prob_neg, prob_pos)) # (num_samples, 2)

    img = load_img(img_path, target_size=img_size)
    img_array = img_to_array(img)
    img_array_uint8 = np.uint8(img_array.copy()) # Simpan versi uint8 untuk visualisasi

    # --- 2. Inisialisasi Penjelas LIME ---
    print(f"  - Inisialisasi LIME Explainer untuk Ensemble...")
    explainer = lime_image.LimeImageExplainer()

    # --- 3. Hasilkan Penjelasan LIME ---
    print(f"  - Menghasilkan penjelasan LIME untuk {os.path.basename(img_path)}...")
    explanation = explainer.explain_instance(
        img_array_uint8, # Gunakan gambar uint8 untuk LIME
        ensemble_predict_fn, # Gunakan fungsi prediksi ensemble
        top_labels=2, # Jelaskan kelas 0 dan 1 (Normal dan Pneumonia)
        hide_color=0, # Warna untuk menyembunyikan superpixels (0,0,0 = hitam)
        num_samples=100 # Jumlah sample lokal (bisa disesuaikan, 100 relatif cepat)
    )

    # --- 4. Ambil Penjelasan untuk Kelas Tertentu ---
    # Ambil prediksi tertinggi dari ensemble
    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0], # Gunakan label tertinggi dari ensemble
        positive_only=False, # Tampilkan fitur positif dan negatif
        num_features=5, # Jumlah superpixels utama untuk dijelaskan
        hide_rest=False # Jangan sembunyikan sisanya
    )

    # --- 5. Visualisasi ---
    print(f"  - Membuat visualisasi LIME untuk {os.path.basename(img_path)}...")
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    # Prediksi ensemble untuk gambar tunggal ini
    ensemble_proba_single = ensemble_predict_fn(np.expand_dims(img_array_uint8, axis=0))[0] # Ambil array pertama
    pred_proba_pneumonia = ensemble_proba_single[1] # Indeks 1 adalah probabilitas pneumonia
    pred_label_idx = 1 if pred_proba_pneumonia > 0.5 else 0
    pred_label_name = class_names[pred_label_idx]

    fig.suptitle(f'Analisis XAI (LIME) Ensemble: Prediksi: {pred_label_name} ({pred_proba_pneumonia:.2f})', fontsize=16)

    axes[0].imshow(img_array_uint8.astype('uint8')); axes[0].set_title('Gambar Asli'); axes[0].axis('off')

    # Visualisasi hasil LIME
    img_bound = mark_boundaries(temp / 255.0, mask) # Gunakan temp yang sudah diwarnai
    axes[1].imshow(img_bound); axes[1].set_title('LIME Explanation (Boundaries)'); axes[1].axis('off')

    # Visualisasi mask LIME saja
    axes[2].imshow(mask, cmap='RdBu', vmin=-1, vmax=1); axes[2].set_title('LIME Mask (Red=Pos, Blue=Neg)'); axes[2].axis('off')

    plt.tight_layout(rect=[0, 0, 1, 0.93])
    plt.savefig(f'lime_analysis_ensemble_{os.path.basename(img_path)}.png', dpi=150)
    plt.show()

    # Kembalikan sesuatu jika perlu (misalnya, informasi ringkas)
    return {'model': 'Ensemble', 'image': os.path.basename(img_path), 'pred_label': pred_label_name, 'pred_proba': pred_proba_pneumonia}


# --- EKSEKUSI UTAMA UNTUK SATU GAMBAR ---

# 1. Definisikan path ke model-model Anda
model_paths = {
    'ResNet50': '/kaggle/input/pneumonia-84/keras/default/1/ResNet50.keras',
    'DenseNet121': '/kaggle/input/pneumonia-92/keras/default/1/densenet121_pneumonia.keras',
    'InceptionV3': '/kaggle/input/pneumonia-82/keras/default/1/inceptionv3_full_model.keras',
}

# 2. Muat model-model
print("Memuat model-model...")
trained_models = load_trained_models(model_paths)
if not trained_models:
    print(" Tidak ada model yang berhasil dimuat. Proses prediksi dan LIME dibatalkan.")
    exit()
else:
    print(f" {len(trained_models)} model berhasil dimuat: {list(trained_models.keys())}\n")

# 3. DEFINISIKAN PATH GAMBAR YANG INGIN DIPROSES
# Sesuai permintaan Anda
image_path = '/kaggle/input/pneumonia/Pneumonia/test/NORMAL/NORMAL2-IM-0302-0001.jpeg'

print(f"  MEMPROSES GAMBAR: {os.path.basename(image_path)}")
print(f"  Path lengkap: {image_path}")

if not os.path.exists(image_path):
    print(f" ERROR: Gambar tidak ditemukan di path yang ditentukan!")
    print(f"   Path yang dicek: {image_path}")
    exit()

# 4. Lakukan prediksi ensemble
print("\n" + "="*50)
print("MELAKUKAN PREDIKSI ENSEMBLE")
print("="*50)
ensemble_weights = [1/len(trained_models)] * len(trained_models) # Gunakan bobot yang sama
class_pred, prob_pred = predict_pneumonia_ensemble(image_path, trained_models, ensemble_weights)

if class_pred is not None and prob_pred is not None:
    print("\n" + "="*50)
    print("PREDIKSI ENSEMBLE SELESAI")
    print("="*50)
    print(f"Gambar: {os.path.basename(image_path)}")
    print(f"Prediksi Ensemble: {class_pred}")
    print(f"Probabilitas Pneumonia: {prob_pred:.4f}")

    # 5. Lakukan analisis LIME ensemble
    print("\n" + "="*50)
    print("MENJALANKAN ANALISIS XAI (LIME) ENSEMBLE")
    print("="*50)
    lime_result = calculate_xai_lime_ensemble(trained_models, ensemble_weights, image_path)

    if lime_result:
        print("\n" + "="*50)
        print("ANALISIS XAI (LIME) SELESAI")
        print("="*50)
        print(f"Gambar: {lime_result['image']}")
        print(f"Prediksi LIME Ensemble: {lime_result['pred_label']}")
        print(f"Probabilitas Pneumonia LIME Ensemble: {lime_result['pred_proba']:.2f}")
    else:
        print("\n Proses analisis XAI (LIME) ensemble gagal.")
else:
    print("\n Proses prediksi ensemble gagal.")
```
